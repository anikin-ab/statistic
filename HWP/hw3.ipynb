## 1.2 
Создать новый признак delta_renovated, который будет содержать разницу в 
годах между годом реновацией дома и годом постройки дома. Если реновации дома не было, то в новом признаке поставьте 0

```python
delta_renovated = []
for i, row in df.iterrows():
    d = row['yr_renovated'] - row['yr_built']
    if d > 0: 
        delta_renovated.append(d)
        
    else: delta_renovated.append(0)
len(delta_renovated)
```

```python
df['delta_renovated'] = delta_renovated
```

```python
df[df['delta_renovated'] > 0].head(10)
```

## Условие 3: 3 задача. Составьте несколько сводных таблиц

### 3.1 Найдите среднюю стоимость домов в зависимости от количества спален
### Отсортируйте от меньшей стоимости к большей

```python
df.pivot_table(index='bedrooms', aggfunc={'price': 'mean'})
```

```python
df['bedrooms'].value_counts()
```

```python
df3 = df.pivot_table(index='bedrooms', aggfunc={'price': 'mean'})
df3.sort_values('price')
```

### 3.2 Найдите минимальную, среднюю и максимальную стоимости домов в зависимости от состояния дома


```python
df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
```

### 3.3 Постройте таблицу с подсчетом количества домов в данных в зависимости от вида на набережную и оценкой вида

```python
df['waterfront'].unique()
```

```python
df['view'].unique()
```

```python
df.groupby(['waterfront', 'view']).agg({'id' : 'count'})
```

```python
df.pivot_table(index='waterfront', columns='view', values='id', aggfunc='count')
```

### 3.4 Каких домов в зависимости от этажности и количества спален больше?

```python
pd.crosstab(index=df['floors'], columns=df['bedrooms'], margins=True, margins_name="Sum")
```

### 3.5 Постройте таблицу с подсчетом медианной стоимости домов в данных в зависимости от состояния дома и оценки дома

```python
df.groupby(['condition', 'grade']).agg({'price' : 'median'})
```

```python
df.groupby(['condition', 'grade']).agg({'price' : 'median'}).unstack('grade').fillna(0)
```
